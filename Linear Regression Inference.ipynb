```python
import pickle
import numpy as np

# Load the trained model and make inference
def predict(input_features):
    with open("linear_regression_model.pkl", "rb") as f:
        model = pickle.load(f)

    # Extract coefficients
    b0 = model["intercept"]
    b1 = model["age_coeff"]
    b2 = model["bmi_coeff"]

    # Input format: [age, bmi]
    age, bmi = input_features

    # Linear regression equation: y = b0 + b1*age + b2*bmi
    prediction = b0 + b1 * age + b2 * bmi

    return prediction

# ✅ Example input for inference
# Format: [age, bmi]
sample_input = [50, 27]

# Run prediction
result = predict(sample_input)

print(f"Predicted SBP: {result}")
```
